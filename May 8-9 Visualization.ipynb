```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("This is title")
```

```python
plt.scatter(x,y,color='red')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("This is title")
```

```python
plt.scatter(x,y,color='red',marker='*')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("This is title")
```

```python
plt.subplot(2,2,1) #2,2 is no rows and no.of coloums you want 1 in which plot u want to plot graph
plt.plot(x,y,'g')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("This is title")
plt.subplot(2,2,4)
plt.scatter(x,y,color='red')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("This is title")
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #([origin coordinates of x, origin coordinates of y , x-axis-displacement, y-axis-displacement])

axes.plot(x,y,'b')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y,'b')
axes.set_xlabel('xaxis')
axes.set_xlabel('yaxis')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'b')
axes.set_xlabel('xaxis')
axes.set_xlabel('yaxis')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'b',label="bigplot")
axes.legend() #to print any thing on graph we use legend
axes.set_xlabel('xaxis')
axes.set_xlabel('yaxis')
axes2.plot(x,y,'red',label='smallplot')
axes2.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'b',label="bigplot")
axes.legend(loc="upper right")
axes.set_xlabel('xaxis')
axes.set_xlabel('yaxis')
axes2.plot(x,y,'red',label='smallplot')
axes2.legend(loc="center")
```

```python
x=np.linspace(0,5,11)
y=x**2
x2=np.linspace(0,10,11)
y2=x**2
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.plot(x2,y2)
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'b',label="bigplot")
axes.legend(loc="upper right")
axes.set_xlabel('xaxis')
axes.set_xlabel('yaxis')
axes2.plot(x,y,'red',label='smallplot')
axes2.scatter(y,x,color='red')

axes2.legend(loc="center")
```

```python
plt.bar(x,y,color="green")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
```

```python
labels=['php','python','java','c']
number=[30,40,15,10]
fig,ax1=plt.subplots()
ax1.pie(number,labels=labels,autopct="%0.2f%%",colors=['orange','blue','pink','yellow'])
ax1.legend(number,loc="upper right")
```
